```python
import matplotlib.pyplot as plt
import seaborn as sns

# Increase the figure size to accommodate all data
plt.figure(figsize=(10, 6))

# Assuming 'merged_corr' contains your dataset
# Adjust other parameters as needed
heatmap = sns.heatmap(df.corr(), annot=True, cmap='Blues', linewidths=.5)

# Set title and labels
plt.title('Correlation Heatmap')

# Rotate the tick labels if needed to prevent overlapping
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels
plt.yticks(rotation=0)  # Keep y-axis labels horizontal

# Adjust font size of annotations if necessary
for _, spine in heatmap.spines.items():
    spine.set_visible(True)

plt.tight_layout()  # Adjust layout to prevent overlap of labels

plt.show()
```
